<h4>Creating the final dataset</h4>

Each data instance consists of 10 consecutive readings from the sensors (ie data from 10 consecutive time stamps in sequence). 
<p>

For example if we have 33 features initially and N number of data, we take 10 rows of data and convert it into features. thus making the features from 33 to 33x10 = 330.


```python
def create_samples(df, sample_size=10, stride=5, max_samples_per_class=1000):
    """
    Create sequential (unflattened) windowed samples from sensor time-series data.

    Parameters
    ----------
    df : pandas.DataFrame
        Column 0  → activity label
        Columns 1..n → sensor features
    sample_size : int
        Number of consecutive rows per window (e.g., 10).
    stride : int
        Step size between consecutive windows (e.g., 5 for 50% overlap).
    max_samples_per_class : int
        Maximum windows to take per activity to keep classes balanced.

    Returns
    -------
    samples_df : pandas.DataFrame
        Two columns:
            • 'label'  : activity name
            • 'window' : NumPy array of shape (sample_size, n_features)
    """

    activity_labels = df.iloc[:, 0].unique()
    labels, windows = [], []

    for activity in activity_labels:
        data = df[df.iloc[:, 0] == activity]
        X = data.iloc[:, 1:].values        # sensor features only
        n_features = X.shape[1]
        length = len(X)
        print(f"Label: {activity}")

        count = 0
        for i in range(0, length - sample_size + 1, stride):
            window = X[i : i + sample_size, :]
            if window.shape == (sample_size, n_features):
                windows.append(window)
                labels.append(activity)
                count += 1
            if count >= max_samples_per_class:
                break

        print(f"  Windows created: {count}")

    # Build the output DataFrame
    samples_df = pd.DataFrame({"label": labels, "window": windows})
    print("Final samples shape:", samples_df.shape)
    return samples_df
```

<h2>Test Train Split</h2>

```python
def zero_shot_split(samples_df, test_ratio=0.2, random_state=41):
    """
    Splits samples into train/test sets by activity (class-based split)
    for Zero-Shot Learning setup.

    Parameters
    ----------
    samples_df : pd.DataFrame
        Output from create_samples(), with columns ['label', 'window'].
    test_ratio : float
        Fraction of unique activities to reserve as unseen/test classes.
    random_state : int
        Random seed for reproducibility.

    Returns
    -------
    train_df, test_df : pd.DataFrame
        DataFrames containing 'label' and 'window' columns for seen and unseen classes.
    seen_classes, unseen_classes : list
        Lists of activity names used in training and testing.
    """

    # All unique activities
    all_activities = samples_df["label"].unique()
    all_activities = np.sort(all_activities)

    # Split activity labels into seen and unseen sets
    seen_classes, unseen_classes = train_test_split(
        all_activities, test_size=test_ratio, random_state=random_state
    )

    # Filter samples by activity
    train_df = samples_df[samples_df["label"].isin(seen_classes)].reset_index(drop=True)
    test_df  = samples_df[samples_df["label"].isin(unseen_classes)].reset_index(drop=True)

    print(f"Seen classes ({len(seen_classes)}): {seen_classes}")
    print(f"Unseen classes ({len(unseen_classes)}): {unseen_classes}")
    print(f"Train samples: {len(train_df)} | Test samples: {len(test_df)}")

    return train_df, test_df, seen_classes.tolist(), unseen_classes.tolist()
```

Data Augmentation

```python
import numpy as np

def jitter(x, sigma=0.01):
    return x + np.random.normal(loc=0, scale=sigma, size=x.shape)

def scaling(x, sigma=0.1):
    factor = np.random.normal(1.0, sigma, (x.shape[1],))
    return x * factor

def time_warp(x, stretch_factor=0.05):
    from scipy.interpolate import interp1d
    orig_steps = np.arange(x.shape[0])
    random_warp = np.linspace(0, x.shape[0]-1, x.shape[0]) * (1 + np.random.uniform(-stretch_factor, stretch_factor))
    f = interp1d(orig_steps, x, axis=0, kind='linear', fill_value="extrapolate")
    return f(np.clip(random_warp, 0, x.shape[0]-1))

def augment_sample(x):
    ops = [jitter, scaling, time_warp]
    aug_x = x.copy()
    for op in ops:
        if np.random.rand() < 0.5:
            aug_x = op(aug_x)
    return aug_x
```

<h2>Model Set up and Training</h2>

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

all_labels = [label_id_to_text[i] for i in sorted(label_id_to_text.keys())]

sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

# Compute embeddings
label_embeddings = sbert_model.encode(all_labels, convert_to_tensor=True)  # (num_classes, S)
label_embeddings = F.normalize(label_embeddings, dim=1)  # L2 normalize
semantic_dim = label_embeddings.shape[1]

# Map label IDs → embedding for easy lookup
id_to_embedding = {k: label_embeddings[i].to(device) for i, k in enumerate(sorted(label_id_to_text.keys()))}
```

```python
class SensorDataset(Dataset):
    def __init__(self, df, id_to_embedding,augment=False):
        """
        df: DataFrame with columns ['label', 'window']
        id_to_embedding: dict mapping label ID → semantic embedding
        """
        self.windows = df['window'].values
        self.labels = df['label'].values
        self.id_to_embedding = id_to_embedding
        self.augment = augment

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        x = self.windows[idx]
        if self.augment:
            x = augment_sample(x)
        x = torch.tensor(self.windows[idx], dtype=torch.float32)  # (T, F)
        y = self.id_to_embedding[self.labels[idx]]               # (semantic_dim,)
        return x, y
```

```python
class SensorTransformerEncoder(nn.Module):
    def __init__(self, n_features, d_model=128, n_heads=4, n_layers=2, dim_feedforward=256, dropout=0.1):
        super().__init__()
        self.input_proj = nn.Linear(n_features, d_model)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=dim_feedforward, dropout=dropout
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)
        self.pool = nn.AdaptiveAvgPool1d(1)

    def forward(self, x):
        x = self.input_proj(x)          # (B, T, D)
        x = x.permute(1, 0, 2)          # (T, B, D)
        x = self.transformer_encoder(x) # (T, B, D)
        x = x.permute(1, 2, 0)          # (B, D, T)
        x = self.pool(x).squeeze(-1)    # (B, D)
        return x

class ProjectionHead(nn.Module):
    def __init__(self, d_model=128, semantic_dim=384):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(d_model, 256),
            nn.ReLU(),
            nn.Linear(256, semantic_dim)
        )

    def forward(self, x):
        x = self.fc(x)
        x = F.normalize(x, dim=-1)
        return x

class ZeroShotHARModel(nn.Module):
    def __init__(self, n_features, d_model=128, semantic_dim=384):
        super().__init__()
        self.encoder = SensorTransformerEncoder(n_features, d_model=d_model)
        self.projector = ProjectionHead(d_model=d_model, semantic_dim=semantic_dim)

    def forward(self, x):
        h = self.encoder(x)
        s_hat = self.projector(h)
        return s_hat
```

Loss Function

```python
def cosine_loss(s_hat, s_true):
    return 1 - F.cosine_similarity(s_hat, s_true, dim=-1).mean()
```

Training Loop

```python
# Hyperparameters
batch_size = 64
lr = 1e-3
epochs = 15
n_features = sample_df['window'][0].shape[1]  # number of sensor features
model = ZeroShotHARModel(n_features, d_model=128, semantic_dim=semantic_dim).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

# Dataset and loader
train_dataset = SensorDataset(train_df, id_to_embedding,augment=True)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Training
model.train()
for epoch in range(epochs):
    total_loss = 0
    for x_batch, y_batch in train_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        s_hat = model(x_batch)
        loss = cosine_loss(s_hat, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * x_batch.size(0)
    print(f"Epoch {epoch+1}/{epochs} | Loss: {total_loss/len(train_dataset):.4f}")
```

Zero-Shot Inference

```python
def predict(model, X_test, unseen_label_ids, id_to_embedding):
    model.eval()
    with torch.no_grad():
        X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
        s_hat = model(X_test_tensor)  # (B, semantic_dim)
        s_hat = F.normalize(s_hat, dim=-1)
        
        # Collect unseen class embeddings
        unseen_embeddings = torch.stack([id_to_embedding[i] for i in unseen_label_ids])  # (num_unseen, S)
        unseen_embeddings = F.normalize(unseen_embeddings, dim=-1)
        
        # Cosine similarity: (B, num_unseen)
        sim = s_hat @ unseen_embeddings.T
        preds_idx = sim.argmax(dim=1)   # index of max similarity
        preds_labels = [unseen_label_ids[i] for i in preds_idx.cpu().numpy()]
    return preds_labels
```

Get Test Accuracy

```python
from sklearn.metrics import accuracy_score

# Prepare test sensor windows and true labels
X_test_array = np.stack(test_df['window'].values)
y_test_array = test_df['label'].values
unseen_label_ids = list(unseen_classes)  # IDs of unseen classes

# Predict
y_pred = predict(model, X_test_array, unseen_label_ids, id_to_embedding)

# Compute zero-shot accuracy
acc = accuracy_score(y_test_array, y_pred)
print(f"Zero-Shot Accuracy on Unseen Classes: {acc*100:.2f}%")
```
